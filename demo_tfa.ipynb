# Teorema Fundamental del Álgebra

Esquema de una prueba topológica del Teorema Fundamental del Álgebra

"Todo polinomio tiene al menos una raíz en el plano complejo."

- Notar: Polinomios son funciones continuas:


$$ \lim_{z \to z_0} {|P(z) - P(z_0)|} = 0$$

- Definir un conjunto $\mathbf{C}_R = \lbrace z \in \mathbb{C} : |z| = R \rbrace $

- Elegir un Polinomio cualquiera $P$, definido como $P(z) = a_n z^n + a_{n-1} z^{n-1} + \ldots a_0$.

- Notar: Si $a_0 = 0$ entonces el origen es una raíz de $P$ ya que $P(0) = a_n 0^n + a_{n-1} 0^{n-1} + \ldots + 0 = 0$.

- En caso de que $a_0 \neq 0$, considerar $\mathbf{D}_R$ la imagen de $\mathbf{C}_R$ por $P$, es decir $$\mathbf{D}_R = P \left( \mathbf{C}_R \right)$$

- Para valores muy pequeños de $R$, $\mathbf{D}_R$ estará muy próximo a $a_0$, es decir

$$\sup_{z \in \mathbf{C}_R} \lbrace |P(z) - a_0| \rbrace \ \xrightarrow[R \to 0]{} 0$$

- Para valores muy grandes de $R$, el primer término siempre va a terminar dominando ya que

$$\lim_{z \in \mathbf{C}_R \\ R \to \infty}
\dfrac{|a_{n-1} z^{n-1} + \ldots + a_0|}{|z^n|} = $$

$$\lim_{ \substack{  z \in \mathbf{C}_R \\ R \to \infty}} |a_{n-1} z^{-1} + \ldots + a_0z^{-n}| = 0 $$

- Entonces para valores muy grandes de $R$ y tomando $$z = R \left( \cos(\theta) + i \sin(\theta) \right) $$ $\mathbf{D}_R$ se aproximará como

$$\mathbf{\tilde{D}}_R =  a_n R^n\left( \cos( n \theta) + i \sin(n\theta) \right)$$

que es la imagen de una circunferencia de radio $R^n$ y para valores suficientemente grandes de $R$, esta circunferencia encierra al origen.

- Esquema de la prueba:

- Para $R \rightarrow 0$  la imagen $\mathbf{D}_R$ se condensa en el punto $a_0 \neq 0$.

- Para $R$ muy grande, la imagen $\mathbf{D}_R$ se aproxima a una circunferencia de radio muy grande.

- Como $P$ es continuo, si $R$ varía de valores muy pequeños a muy grandes, en algún valor intermedio la curva $\mathbf{D}_R$ va a tocar el $0$.
                
- Entonces para algún valor de $R$ se cumple que $$P \left( R\cos(\theta) + i R\sin(\theta) \right) = 0$$

- Es decir que $P$ tendrá al menos una raíz!

- Notar que limitándose a los reales este razonamiento deja de ser válido ya que no hay circunferencias en $\mathbb{R}$.
El equivalente de $\mathbf{C}_R$ en $\mathbb{R}$ es el segmento $-R \leq x \leq R$ de $\mathbb{R}$ para el dominio, y el segmento $\min \left( P \left( \mathbf{C}_R \right) \right) \leq y \leq \max \left( P \left( \mathbf{C}_R \right) \right)$ de $\mathbb{R}$ para el codominio.
Según sean los coeficientes de $P$ se puede dar que $\min \left( P \left( \mathbf{C}_R \right) \right) > 0$  $\forall R \in \mathbb{R}$ (o bien $\max \left( P \left( \mathbf{C}_R \right) \right) < 0$  $\forall R \in \mathbb{R}$) y luego $P$ no se anule en ningún punto de $\mathbb{R}$.

```python
explicación = """

### "Todo polinomio tiene al menos una raíz en el plano complejo."

## Esquema de una prueba topológica:

Los polinomios son funciones continuas, lo que significa que no tienen saltos o interrupciones en su comportamiento.
Definir un conjunto C_R como el conjunto de números complejos cuyo módulo es R. Considerar un polinomio P(z) = a_n*z^n + a_{n-1}*z^{n-1}+...+a_0.
Si el término constante del polinomio a_0 es cero, entonces el origen es una raíz, porque P(0). será cero.
Si a_0 no es cero, observar cómo se comporta la curva violeta P(C_R), la imagen de la curva naranja C_R bajo P a medida que R varía.
Cuando R es muy pequeño, P(C) estará cerca de a_0.
A medida que R se hace muy grande, el término de mayor grado en P dominará a los demás, y P(C_R) empezará a parecerse a una circunferencia cuyo radio crece con R.
Dado que P es continua, al variar R desde valores muy pequeños a muy grandes, en algún punto intermedio P(C_R) debe pasar por el cero (es decir, tocar el punto del origen).
Por lo tanto, en algún punto, P tendrá al menos una raíz en el plano complejo.

Este razonamiento no se sostiene si nos limitamos solo a los números reales, ya que en ese caso no tenemos circunferencias, sino segmentos de línea.
Dependiendo de los coeficientes de P, es posible que el polinomio nunca se anule en los reales, lo que demuestra la importancia de considerar el plano complejo para este teorema.
El gráfico del medio muestra lo que ocurre restringiéndose a los reales. El intervalo del eje y cubierto por P(D), la imagen de D por P, también se muestra en el primer gráfico.
Como son reales, queda marcado sobre el eje real.

## Experimento
1. Escriba el polinomio usando notación de Python ("z al cuadrado" es z**2 y no z^2!!)
2. Varíe R primero con valores pequeños.
3. Observe la curva violeta. ¿Encierra al origen?
4. Varíe R hacia valores más grandes.
5. Observe la curva violeta. ¿Encierra al origen?
6. Variando R, ¿Puede hacer que la curva violeta pase por el origen?

Si lo logró: ¡Felicitaciones, usted encontró una raíz de P!
"""
```

```python
import gradio as gr
import numpy as np

import matplotlib.pyplot as plt

def plot_parametric_curve(r, f):
    """
    Esta función genera y traza la imagen de un círculo por una función en un plano complejo.

    Parámetros:
    - r: Radio del círculo.
    - f: Polinomio en función de z.

    Retorna:
    - fig1: Figura que muestra la imagen del círculo en el plano complejo.
    - fig2: Figura que muestra el polinomio evaluado en los reales.
    - fig3: Figura que muestra el dominio en el plano complejo.
    """

    # Generar datos para la curva paramétrica
    t = np.linspace(0, 2*np.pi, 100)
    z = r*np.cos(t) + 1j*r*np.sin(t)

    w = eval(f)

    z_orig = z
    z = r*np.linspace(-1.1, 1.1, 100)
    t = z
    y = eval(f)
    z = z_orig

    fig1, ax1 = plt.subplots(figsize=(9, 4))
    ax1.plot(w.real, w.imag, color='darkorchid', label='$P(\\mathbf{C})$')
    ax1.set_aspect('equal', 'box')
    plt.xlabel('Re(w)')
    plt.ylabel('Im(w)')
    plt.title('Imagen del Círculo $\\mathbf{C}:|z|=R$ por $P(z)$, $R = '+str(r) +'$')
    plt.plot(y.real, y.imag, 'r', label='$P(\\mathbf{D})$')
    plt.plot(0, 0, '+k', label='$O$')
    plt.plot()
    plt.legend()
    plt.grid('on')
    plt.axis('equal')

    fig2, ax2 = plt.subplots(figsize=(9, 4))
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Polinomio evaluado en los reales')
    ax2.plot(t, 0*t, label='$\\mathbf{D} : -1.1*R<x<1.1*R$')
    plt.plot(t, y.real, 'r', label='$y = Re(P(x))$')
    plt.plot(t, y.imag, ':r', label='$y = Im(P(x))$')
    plt.plot([-r, r], [0, 0],'*g', label='$x = \\pm R$')
    plt.grid('on')
    plt.legend()


    fig3, ax3 = plt.subplots(figsize=(9, 4))
    ax3.plot(t, t*0, label='$\\mathbf{D} : -1.1*R<x<1.1*R$')
    ax3.plot(z.real, z.imag, label='$\\mathbf{C} : |z| = R$')
    plt.xlabel('Re(z)')
    plt.ylabel('Im(z)')
    plt.title('Dominio: Plano $z$')
    plt.plot([-r, r], [0, 0],'*g', label='$x = \\pm R$')
    plt.legend()
    plt.grid('on')
    plt.axis('equal')

    return fig1, fig2, fig3

# Gradio
iface = gr.Interface(
    fn=plot_parametric_curve,
    title="Teorema Fundamental del Álgebra",
    description=explicación,
    inputs=[
        gr.Slider(0.01, 10, label="R"),
        gr.Textbox(label="Polinomio P(z)=  (por ejemplo: z**2 - 5*z + 6) $P(z)=$", value="(z-1)*(z+2)*(z-3)")
    ],
    outputs=[
        gr.Plot(scale=1, label="Imagen del Círculo"),
        gr.Plot(scale=1, label="Polinomio evaluado en los reales"),
        gr.Plot(scale=1, label="Dominio")
    ],
    live=True,
    allow_flagging="never"
)

# Launch
iface.launch(debug=False)
```
